```python
import pandas as pd

def calculate_statistics(df):
    """
  Calculates various statistics for each column in a DataFrame.

  Args:
    df: A pandas DataFrame.

  Returns:
    A pandas DataFrame with the calculated statistics for each column.
  """

    stats_df = pd.DataFrame(columns=["Count", "Missing", "Unique", "Dtype", "Numerical",
                                 "Mode", "Mean", "Min", "25%", "50%", "75%", "Max", "Skew"])

    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            stats_df.loc[col] = [
              df[col].count(),
              df[col].isna().sum(),
              len(df[col].unique()),
              df[col].dtype,
              True,
              df[col].mode()[0],
              df[col].mean(),
              df[col].min(),
              df[col].quantile(0.25),
              df[col].quantile(0.5),
              df[col].quantile(0.75),
              df[col].max(),
              df[col].skew()
          ]
        else:
            stats_df.loc[col] = [
              df[col].count(),
              df[col].isna().sum(),
              len(df[col].unique()),
              df[col].dtype,
              False,
              df[col].mode()[0],
              "", "", "", "", "", "", ""
          ]

    return stats_df
```

```python
# Perform one-way ANOVA
grouped_data = [df['price'][df['neighbourhood group'] == group] for group in df['neighbourhood group'].unique()]

f_statistic, p_value = f_oneway(*grouped_data)

print(f'F-statistic: {f_statistic}')
print(f'P-value: {p_value}')


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in mean prices among neighbourhood groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean prices among neighbourhood groups.")
```

# 6. **Minimum Nights Analysis:**
   - What is the average minimum number of nights required for booking?
   - Does the minimum nights requirement vary based on room type or neighborhood?


```python
average_minimum_nights = df['minimum nights'].mean()
print(f'Average Minimum Nights Required for Booking: {np.round(average_minimum_nights,2)}')
```

```python
min_nights_by_room_type = df.groupby('room type')['minimum nights'].mean()
pd.DataFrame(min_nights_by_room_type)
```

```python
min_nights_by_neighborhood = df.groupby('neighbourhood group')['minimum nights'].mean()
pd.DataFrame(min_nights_by_neighborhood)
```

```python
grouped_data = [df['minimum nights'][df['neighbourhood group'] == group] for group in df['neighbourhood group'].unique()]

f_statistic, p_value = f_oneway(*grouped_data)

print(f'F-statistic: {f_statistic}')
print(f'P-value: {p_value}')


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in mean minimum nights among neighbourhood groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean  minimum nights among neighbourhood groups.")
```

```python
# Perform one-way ANOVA
grouped_data = [df['minimum nights'][df['room type'] == group] for group in df['room type'].unique()]

f_statistic, p_value = f_oneway(*grouped_data)

print(f'F-statistic: {f_statistic}')
print(f'P-value: {p_value}')


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in mean minimum nights among room type.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean minimum nights room type.")
```

- there is a statistically significant difference in the average minimum night stay depending on the neighbourhood group and room type
- the average minimum night stay for one room type may be significantly higher than the average minimum night stay for another room type.
